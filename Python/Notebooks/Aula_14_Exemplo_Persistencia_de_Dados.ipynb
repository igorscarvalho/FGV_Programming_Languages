Persistência de dados

```python
import requests
from bs4 import BeautifulSoup
import string
from collections import Counter
```

```python
requisicao = requests.get('https://en.wikipedia.org/wiki/FIFA_World_Cup')
soup = BeautifulSoup(requisicao.text, "lxml")
texto = soup.text
palavras = texto.split()
```

```python
palavras_maiusculas = [m for m in palavras if m.istitle()]
palavras_maiusculas_limpas = [m.strip(string.punctuation)
                              for m in palavras_maiusculas]
palavras_maiusculas_limpas = [m.strip(string.digits) 
                              for m in palavras_maiusculas_limpas]
palavras_maiusculas_limpas = [m.strip(string.punctuation) 
                              for m in palavras_maiusculas_limpas]
palavras_maiusculas_limpas = [m for m in palavras_maiusculas_limpas if len(m)>1]
```

```python
frequencias = Counter(palavras_maiusculas_limpas)
```

```python
with open('meu_arquivo.txt','w') as f:
    for chave, valor in frequencias.items():
        if valor > 4:
            f.write('A palavra {} aparece {} vezes\n'.format(chave,valor))
```

```python
with open('meu_arquivo.txt','r') as f:
    meu_texto = f.read()
```
